```python
import pandas as pd 
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_excel("dataset_not_normalized.xlsx")
df=df.drop(columns=['cod_exame','Unnamed: 0'])
df
```

```python
parametros = list(df.drop(columns=['diagnostico']).columns)
parametros
```

```python
df.columns
```

```python
parametros=df.drop(columns=['id_paciente','diagnostico']).columns
parametros
```

```python
df_to_plot = pd.melt(frame=df[parametros].reset_index(),id_vars='index',var_name='column_name',value_name='value')
df_to_plot['diagnostico'] = np.where((df_to_plot['index']>=39), 'SE', 'PNES')
df_to_plot
```

```python
df_to_plot.loc[df_to_plot['column_name'] == 'mediaA5']
```

```python
# Create a visualization
sns.relplot(data=df_to_plot,x='index',y='value',hue='diagnostico',style='column_name',palette='plasma')
```
